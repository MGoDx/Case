```python
# Importando as Bibliotecas Necessárias

import pandas as pd
import json
import numpy as np
import zipfile
import os

# Caminho para o arquivo zip e diretório de extração
zip_path = 'data/Data.zip'
extract_dir = 'data/extracted_data/'

# Extraindo os arquivos
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

# Listando os arquivos extraídos para verificar
extracted_files = os.listdir(extract_dir)
extracted_files
```

```python
import pandas as pd

# Carregando o arquivo df_empresas.csv
df_empresas_path = os.path.join(extract_dir, 'df_empresas.csv')
df_empresas = pd.read_csv(df_empresas_path, sep=';')

# Exibindo as primeiras linhas do dataframe para análise inicial
df_empresas.head()
```

```python
import json

# Carregando o arquivo JSON
json_file_path = os.path.join(extract_dir, 'bq-results-20240515-184938-1715799987947.json')

# Carregando o arquivo JSON linha por linha para evitar problemas de múltiplos objetos
processos_data = []
with open(json_file_path, 'r') as json_file:
    for line in json_file:
        try:
            processos_data.append(json.loads(line))
        except json.JSONDecodeError as e:
            print(f"Erro ao decodificar a linha: {e}")

# Exibindo a primeira entrada para entender a estrutura
processos_data[0] if processos_data else "Nenhum dado encontrado"
```

```python
# Carregando e exibindo uma amostra dos outros arquivos CSV para análise

# Caminho dos arquivos
files_to_load = ['empresas_nivel_atividade.csv', 'empresas_porte.csv', 'empresas_saude_tributaria.csv', 'empresas_simples.csv']

# Criar dicionário para armazenar os dataframes
dataframes = {}

# Carregar e visualizar os dados
for file_name in files_to_load:
    file_path = os.path.join(extract_dir, file_name)
    df = pd.read_csv(file_path, sep=';')
    dataframes[file_name] = df.head()  # Armazenar apenas a amostra para análise inicial

dataframes
```

```python
# Carregando os arquivos CSV e JSON novamente para garantir que estão atualizados

# Carregando os arquivos CSV
df_empresas = pd.read_csv(os.path.join(extract_dir, 'df_empresas.csv'), sep=';')
df_nivel_atividade = pd.read_csv(os.path.join(extract_dir, 'empresas_nivel_atividade.csv'), sep=';')
df_porte = pd.read_csv(os.path.join(extract_dir, 'empresas_porte.csv'), sep=';')
df_saude_tributaria = pd.read_csv(os.path.join(extract_dir, 'empresas_saude_tributaria.csv'), sep=';')
df_simples = pd.read_csv(os.path.join(extract_dir, 'empresas_simples.csv'), sep=';')

# Convertendo o JSON para um dataframe
df_processos = pd.DataFrame(processos_data)

# Exibindo a quantidade de registros carregados em cada dataframe
{
    'df_empresas': df_empresas.shape,
    'df_nivel_atividade': df_nivel_atividade.shape,
    'df_porte': df_porte.shape,
    'df_saude_tributaria': df_saude_tributaria.shape,
    'df_simples': df_simples.shape,
    'df_processos': df_processos.shape
}
```

```python
# Função para normalizar o CNPJ (remover espaços em branco e formatar como string)
def normalize_cnpj(cnpj):
    return str(cnpj).zfill(14).replace('.', '').replace('-', '').replace('/', '')

# Aplicando a normalização do CNPJ em todas as tabelas
df_empresas['cnpj'] = df_empresas['cnpj'].apply(normalize_cnpj)
df_nivel_atividade['cnpj'] = df_nivel_atividade['cnpj'].apply(normalize_cnpj)
df_porte['cnpj'] = df_porte['cnpj'].apply(normalize_cnpj)
df_saude_tributaria['cnpj'] = df_saude_tributaria['cnpj'].apply(normalize_cnpj)
df_simples['cnpj'] = df_simples['cnpj'].apply(normalize_cnpj)
df_processos['cnpj'] = df_processos['cnpj'].apply(normalize_cnpj)

# Verificando a normalização em uma amostra
{
    'df_empresas': df_empresas['cnpj'].head(),
    'df_nivel_atividade': df_nivel_atividade['cnpj'].head(),
    'df_porte': df_porte['cnpj'].head(),
    'df_saude_tributaria': df_saude_tributaria['cnpj'].head(),
    'df_simples': df_simples['cnpj'].head(),
    'df_processos': df_processos['cnpj'].head(),
}
```

```python
# Realizando a integração das tabelas usando merge no CNPJ
df_consolidado = df_empresas.merge(df_nivel_atividade, on='cnpj', how='left') \
                            .merge(df_porte, on='cnpj', how='left') \
                            .merge(df_saude_tributaria, on='cnpj', how='left') \
                            .merge(df_simples, on='cnpj', how='left')

# Exibindo as primeiras linhas do dataframe consolidado
df_consolidado.head()
```

```python
# Tratamento de valores nulos
# Substituindo valores nulos nas colunas categóricas por 'Desconhecido'
df_consolidado['nivel_atividade'].fillna('Desconhecido', inplace=True)
df_consolidado['empresa_porte'].fillna('Desconhecido', inplace=True)
df_consolidado['saude_tributaria'].fillna('Desconhecido', inplace=True)
df_consolidado['optante_simples'].fillna('Desconhecido', inplace=True)
df_consolidado['optante_simei'].fillna('Desconhecido', inplace=True)

# Exibindo novamente as primeiras linhas após o tratamento
df_consolidado.head()
```

```python
# Integrando o dataframe consolidado das empresas com os processos jurídicos
df_final = df_consolidado.merge(df_processos, on='cnpj', how='left')

# Exibindo as primeiras linhas do dataframe final
df_final.head()
```

```python
# Separando a coluna "comarca" em duas: "cidade" e "estado"
df_processos[['cidade', 'estado']] = df_processos['comarca'].str.split(' - ', expand=True)

# Verificando as primeiras linhas após a separação
df_processos[['comarca', 'cidade', 'estado']].head()
```

```python
# Função ajustada para separar as partes em duas colunas: "parte_ativa" e "parte_passiva"
def separar_partes(partes):
    try:
        # Trabalhando diretamente com a lista de dicionários
        parte_ativa = next((parte['nomeNormalizadoNeoway'] for parte in partes if parte['polo'] == 'ativo'), None)
        parte_passiva = next((parte['nomeNormalizadoNeoway'] for parte in partes if parte['polo'] == 'passivo'), None)
        return parte_ativa, parte_passiva
    except (ValueError, SyntaxError, TypeError):
        return None, None

# Aplicando a função para criar as colunas "parte_ativa" e "parte_passiva" novamente
df_processos['parte_ativa'], df_processos['parte_passiva'] = zip(*df_processos['partes'].apply(separar_partes))

# Verificando o resultado novamente
df_processos[['partes', 'parte_ativa', 'parte_passiva']].head()
```

```python
# Inspecionando o conteúdo da coluna "assuntosCNJ"
df_processos.head(100)
```

```python
# Função ajustada para extrair os títulos dos assuntos CNJ, lidando com possíveis problemas de chave
def extrair_assuntos(assuntos):
    try:
        # Verificar se a lista de assuntos contém dicionários com a chave 'titulo'
        titulos = [assunto.get('titulo', 'Sem título') for assunto in assuntos if isinstance(assunto, dict)]
        return ', '.join(titulos) if titulos else None
    except (ValueError, TypeError):
        return None

# Aplicando a função para criar uma nova coluna "assuntosCNJ_titulos"
df_processos['assuntosCNJ_titulos'] = df_processos['assuntosCNJ'].apply(extrair_assuntos)

# Verificando as primeiras linhas após o tratamento
df_processos[df_processos['assuntosCNJ_titulos'].notnull()].head(10)
```

```python
# Examinando a coluna "orgaoJulgador" para identificar diferentes padrões
df_processos['orgaoJulgador'].value_counts().head(20)  # Verificar os 20 valores mais comuns para entender os padrões
```

```python
# A análise revelou que a coluna "orgaoJulgador" contém informações estruturadas como "Nome do Órgão - Cidade - Estado".
# Para tornar essas informações mais úteis, podemos separar essa coluna em três novas colunas:
# orgao_nome: O nome do órgão julgador (ex.: "3 VARA CIVEL").
# cidade: A cidade onde o órgão está localizado (ex.: "Sao Paulo").
# estado: A sigla do estado (ex.: "SP").

# Separando a coluna "orgaoJulgador" em três colunas: "orgao_nome", "cidade", e "estado"
df_processos[['orgao_nome', 'cidade_orgao', 'estado_orgao']] = df_processos['orgaoJulgador'].str.rsplit(' - ', n=2, expand=True)

# Exibindo as primeiras linhas após a separação
df_processos[['orgaoJulgador', 'orgao_nome', 'cidade_orgao', 'estado_orgao']].head(10)
```

```python
# Removendo as colunas solicitadas
df_processos = df_processos.drop(columns=['partes', 'assuntosCNJ', 'comarca','orgaoJulgador'])

# Verificando as primeiras linhas após a remoção das colunas
df_processos.head()
```

```python
# Exportando a tabela de empresas (df_consolidado) para um arquivo CSV
empresas_csv_path = os.path.join(extract_dir, 'empresas_consolidado.csv')
df_consolidado.to_csv(empresas_csv_path, index=False)

# Exportando a tabela de processos (df_processos) para um arquivo CSV
processos_csv_path = os.path.join(extract_dir, 'processos.csv')
df_processos.to_csv(processos_csv_path, index=False)

# Verificando se os arquivos foram criados com sucesso
os.path.exists(empresas_csv_path), os.path.exists(processos_csv_path)
```
